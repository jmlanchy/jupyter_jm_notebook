```python
import regex as re
from Bio import SeqIO
from Bio.Seq import Seq

for seq_record in SeqIO.parse("sequence.gb", "genbank"):
    total=0
    matches = re.finditer(r"[AG]T[GATC]{2}GT", str(seq_record.seq), overlapped=True)
    #sequences used for comparison with motif
    #matches = re.finditer(r"GT{1,2}GTG", str(seq_record.seq), overlapped=True)
    for m in matches: 
        base  = m.group() 
        pos   = m.start()
        fin   = m.end()
        total = total + 1
        print("\t" + seq_record.name + "," + base.replace("T","U") + "," + str(pos) + "," + str(fin))     
    print("There are a total of " + str(total) +  " motifs in the " + seq_record.annotations["source"] + " genome.")
```
