```python
from Bio import SeqIO
mydict={}
record = SeqIO.read("KF021605.fasta", "fasta")
for i in range(40,453,1):
    k=0
    mydict['aminoacid{}'.format(i)]=record.seq[i]
    mydict['position{}'.format(i)]=i
    for j in range(1,11):
        mydict['minus{}_{}'.format(i,j)]=record.seq[int(i-j)]
    for k in range(1,11):
        mydict['plus{}_{}'.format(i,k)]=record.seq[i+k]
print(mydict)
```
