```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.Align.Applications import MuscleCommandline
from io import StringIO
from Bio import AlignIO
import sys

################# DNA SEQUENCES EXTRACTION ####################################
f = open("UL75dna.fasta",'a')
for rec in SeqIO.parse("C:/Users/rifty/allHCMVgenomes.gb", "genbank"):
    for feature in rec.features:
        if feature.qualifiers.get('product') == ["envelope glycoprotein H"]:
            beginning = int(feature.location.start)
            ending = int(feature.location.end)
            f.write(">UL75_" + rec.name + "\n")
            f.write(str(rec.seq[beginning:ending].reverse_complement()) + "\n")
f.close()
print("Extraction done.")

################# MUSCLE ALIGNMENT ###############################################
              
in_file = "C:/Users/rifty/UL75dna.fasta"
out_file = "C:/Users/rifty/alignedUL75dna.fasta"
              
muscle_exe = "C:/Users/rifty/muscle3.8.31_i86win32"
cline = MuscleCommandline(muscle_exe, input=in_file, out=out_file)
stdout, stderr = cline(in_file)
align = AlignIO.read(out_file, "fasta")
print("Alignement done.")

################# SINGLE NUCLEOTIDE ITEMIZER ####################################       
        
def itemizer(seq, size):
    return [seq[pos:pos + size] for pos in range(0, len(seq), size)]
numbers=list(range(1,745))                                            # last number is last aligned position + 1.
          
#ff = open("UL75dna.csv",'a')
print("accession",end=","), print(*numbers,sep=",")
for seq_record in SeqIO.parse("alignedUL75dna.fasta","fasta"):        # use a fasta file of the aligned nucleotides.
    seq=itemizer(seq_record.seq,1)                                    # so that gaps are represented by '-' characters.
    print(seq_record.name[5:13],end= ","), print(*seq,sep=",")
    #ff.write("\n")
#ff.close()
print("Last step done. Check the file.")
```
