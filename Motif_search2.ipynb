```python
from Bio import SeqIO
from Bio import motifs
from Bio.Seq import Seq

# possible motifs to search for:
# RNA33, 5-GUAGUGUGUG;RNA34, 5-GUCUGUUGUGUG in Heil et al. Science 2004
# GUGUGU, GUGUGUG, GUUGUGU, GUUGUGUG in 

for seq_record in SeqIO.parse("sequence.gb", "genbank"):
    #print(seq_record.id)
    #print(repr(seq_record.seq))
    #print(len(seq_record))
    instances = [Seq("GTGTG")]
    m = motifs.create(instances)
    total = 0
    for pos, seq in m.instances.search(seq_record.seq):
        total = total + 1
        print("%i : %s" % (pos, seq)) 
    print("There are a total of " + str(total) + " GUGUG motifs in the " + seq_record.id + " genome.")
    print("The " + seq_record.id + " genome is " + str(len(seq_record)) + " nucleotides-long.")
```
