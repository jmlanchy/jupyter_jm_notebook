```python
from Bio import SeqIO
from Bio import motifs
from Bio.Seq import Seq

# possible motifs to search for:
# RNA33, 5-GUAGUGUGUG;RNA34, 5-GUCUGUUGUGUG in Heil et al. 2004, Science 303 (5663):1526-1529.
# GUGUGU, GUGUGUG, GUUGUGU, GUUGUGUG in Forsbach et al. 2008, J Immunol 180:3729-3738.
# I need perhaps another dictionary with the different GU motifs so I can loop thru them

for seq_record in SeqIO.parse("sequence.gb", "genbank"):
    instances = [Seq("GTGTG")]
    m = motifs.create(instances)
    total = 0
    print("List of GUGUG motifs in " + seq_record.annotations["source"] + " genome.\n")
    for pos,seq in m.instances.search(seq_record.seq):
        total = total + 1
        print("%i : %s" %(pos,seq.transcribe()))
        
        #I will save the coordinates of the GU motifs in a CSV file
        
    print("\tThere are a total of " + str(total) + " GUGUG motifs in the " + seq_record.annotations["source"] + " genome.")
    print("\tThe " + seq_record.annotations["source"] + " genome is " + str(len(seq_record)) + " nucleotides-long.\n")
```
