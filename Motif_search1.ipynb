```python
from Bio import SeqIO
from Bio import motifs
from Bio.Seq import Seq 
# RNA33, 5-GUAGUGUGUG;RNA34, 5-GUCUGUUGUGUG from Heil et al. Science 2004
# GUUGUGU

HCMV_TB = Seq("")

instances = [Seq("GT")]
m = motifs.create(instances)
total=0
for pos, seq in m.instances.search(HCMV_TB):
    total = total + 1
#print("%i : %s" % (pos, seq)) 
print("There are a total of "+str(total)+" GU motifs in the HCMV_TB sequence.")

instances = [Seq("GTT")]
m = motifs.create(instances)
total=0
for pos, seq in m.instances.search(HCMV_TB):
    total = total + 1
#print("%i : %s" % (pos, seq)) 
print("There are a total of "+str(total)+" GUU motifs in the HCMV_TB sequence.")

instances = [Seq("GTGT")]
m = motifs.create(instances)
total=0
for pos, seq in m.instances.search(HCMV_TB):
    total = total + 1
#print("%i : %s" % (pos, seq)) 
print("There are a total of "+str(total)+" GUGU motifs in the HCMV_TB sequence.")

instances = [Seq("GTTGT")]
m = motifs.create(instances)
total=0
for pos, seq in m.instances.search(HCMV_TB):
    total = total + 1
#print("%i : %s" % (pos, seq)) 
print("There are a total of "+str(total)+" GUUGU motifs in the HCMV_TB sequence.")

instances = [Seq("GTGTG")]
m = motifs.create(instances)
total=0
for pos, seq in m.instances.search(HCMV_TB):
    total = total + 1
#print("%i : %s" % (pos, seq)) 
print("There are a total of "+str(total)+" GUGUG motifs in the HCMV_TB sequence.")

instances = [Seq("GTTGTGTG")]
m = motifs.create(instances)
total=0
for pos, seq in m.instances.search(HCMV_TB):
    total = total + 1
#print("%i : %s" % (pos, seq)) 
print("There are a total of "+str(total)+" GUUGUGUG motifs in the HCMV_TB sequence.")

instances = [Seq("GTGTGTG")]
m = motifs.create(instances)
total=0
for pos, seq in m.instances.search(HCMV_TB):
    total = total + 1
#print("%i : %s" % (pos, seq)) 
print("There are a total of "+str(total)+" GUGUGUG motifs in the HCMV_TB sequence.")

print("The HCMV_TB genome is "+str(len(HCMV_TB))+" nucleotides-long.")
```
