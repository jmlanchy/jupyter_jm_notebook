```python
import regex as re
import csv
from Bio import SeqIO
from Bio.Seq import Seq

def itemizer(seq, size):
    return [seq[pos:pos + size] for pos in range(0, len(seq), size)]

for seq_record in SeqIO.parse("gHproteins.fasta","fasta"):
    seq=itemizer(seq_record.seq,1)
    #print(seq_record.name[0:4] + "," + seq_record.name[5:13] + ","),
    print(seq_record.name + ","),
    print(*list(range(1,743)),sep=","),
    print("\n"),
    print(*seq,sep=","),
    print("\n")
```
